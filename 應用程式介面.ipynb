```python
import tkinter as tk
from tkinter import messagebox

def calculation():
    try:
        pass
    except:
        messagebox.showerror("Error","請輸入完整矩陣！")

def clear():
    text1.delete(0.0, "end")
    text2.delete(0.0, "end")
    text3.delete(0.0, "end")
    text1.insert('end', 'Start\n')
    text2.insert('end', 'Start\n')
    text3.insert('end', 'Start\n')

def about():
    messagebox.showinfo("Information","輸入矩陣數值,按下'計算'即可求矩陣行列式\n按下清除則會清空矩陣數值！")

window = tk.Tk()
window.iconbitmap('logo.ico')
window.title('Matrix determinant computer')
window.geometry('860x680')


menubar = tk.Menu(window)

filemenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='File', menu=filemenu)
filemenu.add_command(label='Exit', command=window.destroy)

editmenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='Edit', menu=editmenu)
editmenu.add_command(label='calculation', command=calculation)
editmenu.add_command(label='clear', command=clear) 

helpmenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label='Help', menu=helpmenu)
helpmenu.add_command(label='about', command=about) 

window.config(menu=menubar)


frame1 = tk.Frame(window)
frame1.pack(side="left", fill='both',expand=True)
frame2 = tk.Frame(window)
frame2.pack(side="top", fill='both',expand=True)
frame3 = tk.Frame(window)
frame3.pack(side="bottom", fill='both',expand=True)


sbar1 = tk.Scrollbar(frame1)
text1 = tk.Text(frame1, wrap='word')
text1.insert('end', 'Start\n')
sbar1.pack(side='right', fill='both',expand=False)
text1.pack(side='left', fill='both',expand=True)
sbar1["command"] = text1.yview
text1["yscrollcommand"] = sbar1.set


sbar2 = tk.Scrollbar(frame2, orient='horizontal')
text2 = tk.Text(frame2, wrap='none')
text2.insert('end', 'Start\n')
sbar2.pack(side='bottom', fill='both',expand=False)
text2.pack(side='left', fill='both',expand=True)
sbar2["command"] =text2.xview
text2["xscrollcommand"] =sbar2.set


sbar3 = tk.Scrollbar(frame3)
text3 = tk.Text(frame3, wrap='word')
text3.insert('end', 'Start\n')
sbar3.pack(side='right', fill='both',expand=False)
text3.pack(side='left', fill='both',expand=True)
sbar3["command"] = text3.yview
text3["yscrollcommand"] = sbar3.set


window.mainloop()
```
