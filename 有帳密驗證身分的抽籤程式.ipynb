```python
import tkinter as tk
from tkinter import messagebox
import random

def ok():
    global confirm
    if entry_account.get()=="Python" and entry_password.get()=="1989":
        verification.destroy()
        confirm = True
    else:
        entry_account.delete(0, "end")
        entry_password.delete(0, "end")
        messagebox.showerror("Information","Incorrect account password！")

def draw_straws():
    members = ["A", "B", "C", "D", "E"]
    label.configure(text=random.choice(members))

confirm = False
    
verification = tk.Tk()
verification.title('Draw straws')
verification.geometry('600x400')
verification.resizable(width=False, height=False)

label_info = tk.Label(font="Helvetica 15 bold", anchor="c", text="Please enter your account password")
label_info.config(width=30 ,height=3)
label_info.grid(column=0, row=0)

label_account = tk.Label(font="Helvetica 15 bold", anchor="c", text="account")
label_account.config(width=15 ,height=3)
label_account.grid(column=0, row=1)

label_password = tk.Label(font="Helvetica 15 bold", anchor="c", text="password")
label_password.config(width=15 ,height=3)
label_password.grid(column=0, row=2)

entry_account = tk.Entry(font="JhengHei 15", borderwidth=3)
entry_account.grid(column=1, row=1)

entry_password = tk.Entry(font="JhengHei 15", borderwidth=3, show='*')
entry_password.grid(column=1, row=2)

button = tk.Button(text="OK", command=ok)
button.config(width=15 ,height=3)
button.grid(column=0, row=3)

verification.mainloop()
    
#============================================================   

if confirm:
    
    root = tk.Tk()
    root.title('Draw straws')
    root.geometry('400x200')
    messagebox.showinfo("Information","Welcome to the automatic lottery program！")
    
    label = tk.Label(bg="#FFE873", font="Helvetica 15 bold", anchor="c", text="who?")
    label.config(width=15 ,height=3)

    button = tk.Button(text="calculation", command=draw_straws)
    button.config(width=15 ,height=3)

    label.pack()
    button.pack()

    root.mainloop()
```
